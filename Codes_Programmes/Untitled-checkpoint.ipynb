```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider, Button

# Paramètres initiaux
ma = 0.5  # Indice de modulation
fm = 1    # Fréquence du signal modulant (Hz)
fp = 10   # Fréquence de la porteuse (Hz)
Up = 1    # Amplitude de la porteuse

# Fonction pour générer le signal modulé
def generate_am_signal(t, ma, fm, fp, Up):
    carrier = Up * np.cos(2 * np.pi * fp * t)
    modulating_signal = np.cos(2 * np.pi * fm * t)
    am_signal = (1 + ma * modulating_signal) * carrier
    return am_signal, modulating_signal, carrier

# Fonction pour mettre à jour les graphiques
def update(val):
    ma = sma.val
    fm = sfm.val
    fp = sfp.val
    Up = sup.val
    am_signal, modulating_signal, carrier = generate_am_signal(t, ma, fm, fp, Up)
    l1.set_ydata(modulating_signal)
    l2.set_ydata(carrier)
    l3.set_ydata(am_signal)
    ax1.set_ylim(-2, 2)
    ax2.set_ylim(-2, 2)
    ax3.set_ylim(-2, 2)
    fig.canvas.draw_idle()

# Création de la figure et des axes
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 8))
plt.subplots_adjust(left=0.1, bottom=0.3)

# Temps
t = np.linspace(0, 1, 1000, endpoint=False)

# Génération des signaux initiaux
am_signal, modulating_signal, carrier = generate_am_signal(t, ma, fm, fp, Up)

# Tracé des signaux
l1, = ax1.plot(t, modulating_signal, lw=2, color='red')
ax1.set_title('Signal modulant')
ax1.grid(True)  # Ajout de la grille
l2, = ax2.plot(t, carrier, lw=2, color='blue')
ax2.set_title('Porteuse')
ax2.grid(True)  # Ajout de la grille
l3, = ax3.plot(t, am_signal, lw=2, color='green')
ax3.set_title('Signal modulé')
ax3.grid(True)  # Ajout de la grille

# Ajout des sliders
axcolor = 'lightgoldenrodyellow'
axma = plt.axes([0.1, 0.2, 0.8, 0.03], facecolor=axcolor)
axfm = plt.axes([0.1, 0.15, 0.8, 0.03], facecolor=axcolor)
axfp = plt.axes([0.1, 0.1, 0.8, 0.03], facecolor=axcolor)
axup = plt.axes([0.1, 0.05, 0.8, 0.03], facecolor=axcolor)

sma = Slider(axma, 'ma', 0.1, 1.0, valinit=ma)
sfm = Slider(axfm, 'fm', 0.1, 5.0, valinit=fm)
sfp = Slider(axfp, 'fp', 1.0, 20.0, valinit=fp)
sup = Slider(axup, 'Up', 0.1, 2.0, valinit=Up)

# Mise à jour des graphiques lorsque les sliders sont modifiés
sma.on_changed(update)
sfm.on_changed(update)
sfp.on_changed(update)
sup.on_changed(update)

# Ajout d'un bouton pour réinitialiser les paramètres
resetax = plt.axes([0.8, 0.025, 0.1, 0.04])
button = Button(resetax, 'Reset', color=axcolor, hovercolor='0.975')

def reset(event):
    sma.reset()
    sfm.reset()
    sfp.reset()
    sup.reset()

button.on_clicked(reset)

plt.show()
```
