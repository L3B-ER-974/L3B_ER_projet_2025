```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
import ipywidgets as widgets
from IPython.display import display

# Activer l'intgration interactive de Matplotlib avec Jupyter
%matplotlib notebook

# Paramtres initiaux
ma = 0.5  # Indice de modulation
fm = 1    # Frquence du signal modulant (Hz)
fp = 10   # Frquence de la porteuse (Hz)
Up = 1    # Amplitude de la porteuse

# Fonction pour gnrer le signal modul
def generate_am_signal(t, ma, fm, fp, Up):
    carrier = Up * np.cos(2 * np.pi * fp * t)
    modulating_signal = np.cos(2 * np.pi * fm * t)
    am_signal = (1 + ma * modulating_signal) * carrier
    return am_signal, modulating_signal, carrier

# Fonction pour mettre  jour les graphiques
def update(ma, fm, fp, Up):
    am_signal, modulating_signal, carrier = generate_am_signal(t, ma, fm, fp, Up)
    l1.set_ydata(modulating_signal)
    l2.set_ydata(carrier)
    l3.set_ydata(am_signal)
    ax1.set_ylim(-2, 2)
    ax2.set_ylim(-2, 2)
    ax3.set_ylim(-2, 2)
    fig.canvas.draw_idle()

# Cration de la figure et des axes
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 8))
plt.subplots_adjust(left=0.1, bottom=0.3)

# Temps
t = np.linspace(0, 1, 1000, endpoint=False)

# Gnration des signaux initiaux
am_signal, modulating_signal, carrier = generate_am_signal(t, ma, fm, fp, Up)

# Trac des signaux
l1, = ax1.plot(t, modulating_signal, lw=2, color='red')
ax1.set_title('Signal modulant')
ax1.grid(True)
l2, = ax2.plot(t, carrier, lw=2, color='blue')
ax2.set_title('Porteuse')
ax2.grid(True)
l3, = ax3.plot(t, am_signal, lw=2, color='green')
ax3.set_title('Signal modul')
ax3.grid(True)

# Ajout des sliders interactifs
ma_slider = FloatSlider(min=0.1, max=1.0, step=0.1, value=ma, description='ma')
fm_slider = FloatSlider(min=0.1, max=5.0, step=0.1, value=fm, description='fm')
fp_slider = FloatSlider(min=1.0, max=20.0, step=1.0, value=fp, description='fp')
Up_slider = FloatSlider(min=0.1, max=2.0, step=0.1, value=Up, description='Up')

# Fonction pour mettre  jour les graphiques lorsque les sliders sont modifis
def on_value_change(change):
    update(ma_slider.value, fm_slider.value, fp_slider.value, Up_slider.value)

ma_slider.observe(on_value_change, names='value')
fm_slider.observe(on_value_change, names='value')
fp_slider.observe(on_value_change, names='value')
Up_slider.observe(on_value_change, names='value')

# Affichage des sliders
display(ma_slider, fm_slider, fp_slider, Up_slider)

plt.show()
```
